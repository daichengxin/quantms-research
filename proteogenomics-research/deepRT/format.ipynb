```python
import pandas as pd
```

```python
input_psm_table = r"vali_out.tsv"
validate_out = pd.read_table(input_psm_table, header = 0, sep = "\t", index_col=0)
```

```python
validate_out
```

```python
#Quantity statistics of whether to use spectrumAI validation
validate_out["status"].value_counts()
```

```python
#Quantity statistics that are validated by spectrumAI
validate_out["flanking_ions_support"].value_counts()
```

```python
#Substitution strategy
def replace_mod(s):
    #Method1
    # s = s.replace("C(Carbamidomethyl)","1").replace("K(TMT6plex)","2").replace("M(Oxidation)","3").replace(".(Acetyl)","").replace(".(TMT6plex)","").replace("(TMT6plex)","")

    #Method2
    s = s.replace("C(Carbamidomethyl)","1").replace("K(TMT6plex)","2").replace("M(Oxidation)","3").replace(".(Acetyl)","4").replace(".(TMT6plex)","5").replace("(TMT6plex)","6")
    return s

validate_out.loc[:,"sequence"] = validate_out.apply(lambda x : replace_mod(x["opt_global_cv_MS:1000889_peptidoform_sequence"]),axis = 1)
validate_out.loc[:,"RT"] = validate_out.apply(lambda x : x["retention_time"],axis = 1)
```

```python
validate_out
```

```python
filter_out = validate_out[(validate_out["position"]=="non-canonical") | (validate_out["position"]=="canonical") | (validate_out["flanking_ions_support"] == "YES") ]
```

```python
filter_out
```

```python
can = filter_out[filter_out["position"]=="canonical"]
```

```python
non = filter_out[filter_out["position"]!="canonical"]
```

```python
can.to_csv(r"can_to_rt.txt", sep = "\t", index = 0)
non.to_csv(r"non_to_rt.txt", sep = "\t", index = 0)
```
